# Visualization of training and test curves with different optimizers
I compare the performace of AdaFilter-SGD optimizer and 7 other optimizers ('SGD','Adam','RAdam','AdamW','MSVAG','SophiaG','Lion').

```python
import os
#matplotlib notebook
import matplotlib.pyplot as plt
import torch
import numpy as np

params = {'axes.labelsize': 20,
          'axes.titlesize': 20,
         }
plt.rcParams.update(params)
```

```python
def get_data(names):
    folder_path = '/classification_cifar10/curve'
    paths = [os.path.join(folder_path, name) for name in names]
    return {name: torch.load(fp) for name, fp in zip(names, paths)}

def plot(names, curve_type='train', labels=None, loc='upper left', ylim=None):
    plt.figure()
    if ylim:
        plt.ylim(ylim[0], ylim[1])
    curve_data = get_data(names)
    for i, label in zip(curve_data.keys(), labels):
        acc = np.array(curve_data[i]['{}_acc'.format(curve_type.lower())])
        if label == 'AdaFilter-SGD':
            plt.plot(acc, '-', label=label)
        else:
            plt.plot(acc, '--', label=label)
    
    plt.grid()
    plt.legend(fontsize=14, loc=loc)
    plt.title('{} accuracy ~ Training epoch'.format(curve_type))
    plt.xlabel('Training Epoch')
    plt.ylabel('Accuracy')
    plt.show()
```

# ResNet
Plot the training and test curves for all optimizers in one plot. "names" is a list containing the log files in "/curve" folder, "labels" is the corresponding legends for different optimizers. Note that "names" and "labels" must match (log for the i-th element in "labels" in the i-th element in "names")

```python
names = ['resnet-sgdf-lr0.5-betas0.9-0.999-wdecay0.0005-eps1e-08-run0-resetFalse',
         'resnet-sgd-lr0.1-momentum0.9-wdecay0.0005-run0-resetFalse',
         'resnet-adam-lr0.001-betas0.9-0.999-wdecay0.0005-eps1e-08-run0-resetFalse',
         'resnet-radam-lr0.001-betas0.9-0.999-wdecay0.0005-eps1e-08-run0-resetFalse',
         'resnet-adamw-lr0.001-betas0.9-0.999-wdecay0.0005-eps1e-08-run0-resetFalse',
         'resnet-msvag-lr0.1-betas0.9-0.999-wdecay0.0005-eps1e-08-run0-resetFalse',
         'resnet-sophia-lr0.0001-betas0.965-0.99-rho0.04-wdecay0.1-run0-resetFalse',
         'resnet-lion-lr2e-05-betas0.9-0.99-wdecay0.1-run0-resetFalse',    
]
labels = ['AdaFilter-SGD',
          'SGD',
          'Adam',
          'RAdam',
          'AdamW',
          'MSVAG',
          'SophiaG',
          'Lion',
        ]
plot(names, 'Train', labels, ylim=(80,101))
plot(names, 'Test', labels, ylim = (84,96))     
```
